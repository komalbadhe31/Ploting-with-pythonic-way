# Multiple Plots one above the other

```python
# Generate the DATA
x = np.arange(0,10,0.2)
y = np.sin(x)
z = np.cos(x)

# Generate the Figure and the Axes
fig , axs = plt.subplots(nrows= 2, ncols= 1)

# plot the SINE curve and the LABEL the ORDINATE of the FIRST graph
axs[0].plot(x,y)
axs[0].set_ylabel('SINE')

# plot the COSINE curve and label the ORDINATE of the SECOND graph
axs[1].plot(x,z)
axs[1].set_ylabel('COSINE')

plt.show()
```

## Gaussian (Normal)

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x-mu,2.)/(2*np.power(sig,2.0)))


x_values = np.linspace(-3,3,5)

plt.plot(x_values, gaussian(x_values,0,1))
```

# custim defined Univariate Normal Density function

```python
def univariate_normal(x, mean , variance):
    """ pdf of the univariate normal distribution (Continuous) """
    return ( (1./np.sqrt(2 * np.pi * variance)) * np.exp(-( x - mean)**2 / (2 * variance)))
                                                                           
```

# Plotting simple univariate Normal

```python
x =  np.linspace(-10,20, 300)  # populate the X axis

fig = plt.figure(figsize=(10,7))  # controlling the size of the plot

plt.plot(x, univariate_normal(x, mean=  0 , variance = 1),
         label="$\mathcal{N}(0,1)$")
plt.plot(x, univariate_normal(x, mean = 0 , variance = 3),
         label="$\mathcal{N}(0,3)$")

plt.legend()
plt.show()

```

# plot Normal distributions with same variance but different means


```python
x =  np.linspace(-10,20, 300)  # populate the X axis

fig = plt.figure(figsize=(10,7))  # controlling the size of the plot

plt.plot(x, univariate_normal(x, mean=  0 , variance = 3),
         label="$\mathcal{N}(0,3)$")
plt.plot(x, univariate_normal(x, mean = 5 , variance = 3),
         label="$\mathcal{N}(5,3)$")

plt.legend()
plt.show()
```

# Cumulative normal distribution function

```python
from scipy.stats import norm
```

```python
fig , ax = plt.subplots(figsize=(10,6))

# code for the total normal pdf
x = np.arange(-4 ,4 , 0.001)
ax.plot(x, norm.pdf(x))
ax.set_title("Cumulative noraml distribution")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x')
ax.grid(True)

# fill some area beteen -4 and say 1
px = np.arange(-4 , 1, 0.01)
ax.fill_between(px, norm.pdf(px), alpha= 0.5,color ='g')

# to display internal text
ax.text(-1,0.1, "cdf(x)",fontsize = 20)

plt.show()
```
